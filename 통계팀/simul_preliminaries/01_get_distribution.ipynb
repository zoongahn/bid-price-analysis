### 2. 4개 구간 선택 및 합성곱 (Convolution)

친구가 말한 “4개의 구간을 뽑고, 각 구간에서 랜덤한 숫자를 뽑아 다 더한 뒤 4로 나눈다”는 절차는

- 우선, 각 선택된 구간에 대해 균등분포(pdf)가 주어지고,
- 그 4개의 독립(혹은 선택이 독립적인) random 변수의 합의 분포는 **합성곱**을 통해 구할 수 있음.

즉, 단위가 되는 분포는 합성곱의 형태로 구해  추출할 수 있음.

**주의:** 만약 구간 선택이 “복원 추출 없이” 이루어진다면 약간의 상관관계가 생기겠으나, 전체적으로 다루시려는 것은 각 구간의 pdf를 합성곱하는 방식으로 접근하는 것임. (이 경우, 각 random 변수의 **한계분포**는 위의 혼합분포와 동일합니다.)

### 2. 각 조합에서의 분포 계산 – Piecewise Polynomial Convolution

1단계: 전체 1365개 조합을 5가지 케이스(4/0, 3/1, 2/2, 1/3, 0/4)로 분리하여 각 케이스마다 동일한 구조(평균 ± 대칭 형태)를 가진 pdf를 구한다.

2단계: 각 케이스별 pdf는 각 조합마다 piecewise polynomial으로 나타내고,

3단계: 각 케이스의 pdf를 평행이동(shift)하여 올바른 지원 상에서 정렬한 후,

4단계: 조합 수에 따른 가중치를 적용해 전체 pdf로 결합!

5단계: 마지막으로, 이 전체 pdf를 이용하여 cdf 및 구간 확률 함수를 구현

```python
import itertools
import numpy as np
from math import comb

# 다항식을 리스트로 표현: poly = [a0, a1, a2, ...]는 a0 + a1*x + a2*x**2 + ... 를 의미함

def eval_poly(poly, x):
    """다항식 poly를 x에서 평가"""
    return sum(c * (x**i) for i, c in enumerate(poly))

def poly_integral(poly):
    """다항식 poly의 부정적분(상수항 0)을 반환. [a0, a1, ...] -> [0, a0, a1/2, ...]"""
    return [0] + [poly[i] / (i+1) for i in range(len(poly))]

def poly_add(poly1, poly2):
    """두 다항식의 덧셈"""
    n = max(len(poly1), len(poly2))
    res = [0]*n
    for i in range(n):
        a = poly1[i] if i < len(poly1) else 0
        b = poly2[i] if i < len(poly2) else 0
        res[i] = a + b
    return res

def poly_mul(poly1, poly2):
    """두 다항식의 곱 (합성곱)"""
    res = [0]*(len(poly1) + len(poly2) - 1)
    for i, a in enumerate(poly1):
        for j, b in enumerate(poly2):
            res[i+j] += a * b
    return res

# 각 구간(Uniform Distn)을 표현하는 클래스 또는 튜플
# 여기서는 (start, end, poly)로 표현하며, poly는 해당 구간에서의 pdf 다항식.
class Segment:
    def __init__(self, start, end, poly):
        self.start = start
        self.end = end
        self.poly = poly

    def __repr__(self):
        return f"Segment([{self.start}, {self.end}], poly={self.poly})"

# 1. 좌측 구간 8개와 우측 구간 7개 생성
left_intervals = []
delta_left = 3/8  # 구간 길이
for i in range(8):
    a = -3 + i * delta_left
    b = -3 + (i+1) * delta_left # 균등분포 pdf: 1/(delta_left) * (1/15)  -- (구간 선택 확률 1/15)
    c = 1 / (15 * delta_left)
    left_intervals.append(Segment(a, b, [c]))

right_intervals = []
delta_right = 3/7
for i in range(7):
    a = 0 + i * delta_right
    b = 0 + (i+1) * delta_right
    c = 1 / (15 * delta_right)
    right_intervals.append(Segment(a, b, [c]))

# 2. 합성곱 함수 (두 piecewise 함수의 합성곱)
# 여기서는 두 Segment의 합성곱을 uniform 구간끼리의 경우를 예로 들겠습니다.
def convolve_two_uniform(seg1, seg2):
    """
    seg1: uniform pdf on [a1, b1] with constant density c1 (poly = [c1])
    seg2: uniform pdf on [a2, b2] with constant density c2 (poly = [c2])
    반환: seg_list, 각 원소는 (start, end, poly) 형태의 Segment,
         합성곱 결과는 [a1+a2, b1+b2]에서 정의되는 piecewise linear function.
    """
    a1, b1, c1 = seg1.start, seg1.end, seg1.poly[0]
    a2, b2, c2 = seg2.start, seg2.end, seg2.poly[0]
    segments = []
    
    z0 = a1 + a2
    z1 = a1 + b2
    z2 = b1 + a2
    z3 = b1 + b2
    
    # 첫 구간: z in [z0, z1]: 선형 증가
    # f(z) = c1*c2 * (z - (a1+a2))
    # 다항식 표현: slope = c1*c2, intercept = -c1*c2*(a1+a2)
    segments.append(Segment(z0, z1, [-c1*c2*(a1+a2), c1*c2]))
    
    # 중간 구간: [z1, z2] (존재할 경우): 상수
    if z1 < z2:
        segments.append(Segment(z1, z2, [c1*c2*(b2 - a2)]))
    
    # 마지막 구간: [z2, z3]: 선형 감소
    # f(z) = c1*c2 * (b1+b2 - z)
    segments.append(Segment(z2, z3, [c1*c2*(b1+b2), -c1*c2]))
    
    return segments

# 3. 일반적인 합성곱 (piecewise 함수와 uniform의 합성곱)
def convolve_piecewise_with_uniform(piecewise_segments, uniform_seg):
    """
    piecewise_segments: 리스트, 각 원소는 Segment (정의역: [a, b], 다항식 poly)
    uniform_seg: 단일 uniform Segment (poly는 상수 함수)
    반환: 두 함수의 합성곱 결과를 나타내는 새로운 Segment 리스트.
    (각 Segment는 [new_a, new_b]와 다항식으로 표현)
    
    아이디어:
    각 원소 seg in piecewise_segments와 uniform_seg에 대해,
    합성곱 결과는
      I(z) = ∫_{x = max(seg.start, z - uniform_seg.end)}^{min(seg.end, z - uniform_seg.start)} seg.poly(x) * uniform_const dx
    이며, 구간별로 적분 상한/하한이 선형 함수로 변할 때 break point에서 분할.
    break point 주의!!!!! 주의!!!!! 주의!!!!!
    """
    new_segments = []
    # 각 seg와 uniform_seg에 대해
    for seg in piecewise_segments:
        # 구간: seg는 [a1, b1] with 다항식 poly1, uniform_seg는 [a2, b2] with 상수 c2.
        a1, b1, poly1 = seg.start, seg.end, seg.poly
        a2, b2, c2 = uniform_seg.start, uniform_seg.end, uniform_seg.poly[0]
        # breakpoints: z0 = a1+a2, z1 = a1+b2, z2 = b1+a2, z3 = b1+b2
        z0 = a1 + a2
        z1 = a1 + b2
        z2 = b1 + a2
        z3 = b1 + b2
        # 아래는 두 부분의 경우(예: poly1가 상수인 경우와 일반적인 경우를 구분)로 나누어
        # 부정적분된 poly1를 활용해 구간별 적분 값을 구하는 방식의 뼈대 코드입니다.
        F_poly1 = poly_integral(poly1)  # antiderivative의 다항식 계수
        
        # 각 구간에서의 합성곱 결과는:
        # 예1: z in [z0, z1]: I(z) = c2 * (F_poly1(z - a2) - F_poly1(a1))
        # 예2: z in [z1, z2]: I(z) = c2 * (F_poly1(b1) - F_poly1(a1))
        # 예3: z in [z2, z3]: I(z) = c2 * (F_poly1(b1) - F_poly1(z - b2))
        # (실제 다항식 계산은 poly 연산으로 수행해야 함)
        
        # 여기서는 간단히 각 구간의 다항식 계수를 직접 계산하는 대신,
        # “세그먼트 생성”의 아이디어만 표현.
        # 실제 구현 시에는 sympy 또는 직접 다항식 연산을 구현할 수 있습니다.
        
        # 구간 1
        new_segments.append(Segment(z0, z1, [None]))  # TODO: 다항식 계수 계산
        # 구간 2 (존재하면)
        if z1 < z2:
            new_segments.append(Segment(z1, z2, [None]))
        # 구간 3
        new_segments.append(Segment(z2, z3, [None]))
    # 여러 세그먼트가 겹치는 경우 병합하는 처리가 필요함.
    return new_segments

# 4. 4개 구간에 대해 반복 합성곱 수행 함수
def convolve_segments_list(segments_list):
    """
    segments_list: 리스트, 각 원소가 Segment (선형 혹은 다항식)
    4개 구간에 대해, for문을 통해 합성곱을 수행
    (초기값은 첫 번째 seg, 이후에 나머지 seg와 convolve_piecewise_with_uniform로 합성곱)
    """
    # 초기 piecewise pdf: 첫 번째 선택된 구간의 pdf
    piecewise = [segments_list[0]]
    for seg in segments_list[1:]:
        # seg는 uniform 분포이므로, 아래와 같이 convolve_piecewise_with_uniform 적용
        piecewise = convolve_piecewise_with_uniform(piecewise, seg)
        # 필요 시, 병합 함수 호출: piecewise = merge_segments(piecewise)
    return piecewise

# 5. 각 케이스별로 조합을 구하고, 그에 따른 pdf를 구하는 과정
def compute_case_pdf(left_count, right_count):
    """
    left_count: 좌측 구간에서 선택할 개수
    right_count: 우측 구간에서 선택할 개수
    4개 선택이므로 left_count + right_count == 4.
    반환: 이 케이스에 대해 모든 조합(pdf들의 합)과, 조합 수.
    """
    case_pdf = None  # 누적 pdf (piecewise 형태)
    comb_count = 0
    # 모든 조합 순회
    for left_choice in itertools.combinations(left_intervals, left_count):
        for right_choice in itertools.combinations(right_intervals, right_count):
            comb_count += 1
            selected = list(left_choice) + list(right_choice)
            # 4개 구간의 pdf를 합성곱하여 하나의 piecewise pdf로 구함
            pdf_piece = convolve_segments_list(selected)
            # 평균을 구하기 위해 X/4로 스케일 조정: f_Y(y) = 4 f_X(4y)
            # 이를 위한 평행이동 및 스케일링 처리 함수 필요 (여기서는 생략)
            # 최종 pdf_piece에 대해 case_pdf에 누적합산 (동일 구간의 경우 합산)
            if case_pdf is None:
                case_pdf = pdf_piece
            else:
                # 두 piecewise pdf의 합산 (동일 구간에서 계수 합산)
                # merge_piecewise_pdf(case_pdf, pdf_piece) 함수로 처리
                case_pdf = merge_piecewise_pdf(case_pdf, pdf_piece)
    return case_pdf, comb_count

# 6. 5가지 케이스에 대해 pdf 계산 후, 가중합으로 전체 pdf 구성
cases = [(4,0), (3,1), (2,2), (1,3), (0,4)]
total_combinations = sum(comb_count for _, comb_count in 
                         [compute_case_pdf(lc, rc) for lc, rc in cases])
# 실제로는 위와 같이 각 케이스의 조합 수(C(n,k))를 곱한 합이 1365임을 확인할 수 있음.

# 각 케이스별 pdf와 조합 수를 저장
case_results = []
for left_count, right_count in cases:
    pdf_case, count_case = compute_case_pdf(left_count, right_count)
    case_results.append((pdf_case, count_case))
    
# 7. 전체 pdf는 각 케이스 pdf에 조합수 가중치를 곱해 더한 후, 전체 합(1365)으로 나눈 값
# 전체 pdf = sum( count_case * shifted_pdf_case ) / 1365
# 여기서 shifted_pdf_case는 각 케이스에 대해 적절한 평행이동(shift)와 스케일링을 적용한 결과.
# 이 과정도 merge_piecewise_pdf와 shift_piecewise_pdf 함수를 통해 구현합니다.

# 8. 최종 cdf 및 구간 확률 함수 작성
def compute_cdf(x, piecewise_pdf):
    """
    piecewise_pdf: 리스트, 각 원소가 Segment(start, end, poly)
    x: 평가 위치
    반환: cdf(x)
    """
    total = 0
    for seg in piecewise_pdf:
        if x >= seg.end:
            # 구간 전체 적분: ∫_seg.start^seg.end poly(x) dx
            F = eval_integral_poly(seg.poly, seg.end) - eval_integral_poly(seg.poly, seg.start)
            total += F
        elif x > seg.start:
            F = eval_integral_poly(seg.poly, x) - eval_integral_poly(seg.poly, seg.start)
            total += F
    return total

def probability_between(a, b, piecewise_pdf):
    return compute_cdf(b, piecewise_pdf) - compute_cdf(a, piecewise_pdf)

# eval_integral_poly 함수는 주어진 다항식의 부정적분을 평가하는 함수로, 필요 시 구현합니다.
# 예:
def eval_integral_poly(poly, x):
    """다항식 poly의 부정적분을 x에서 평가 (적분 상수는 0로 가정)"""
    integ_poly = poly_integral(poly)
    return eval_poly(integ_poly, x)
```

### 1. 다항식 적분 등 함수 구현.

```python
import math
from math import comb

def eval_poly(poly, x):
    """다항식 poly를 x에서 평가 (poly = [a0, a1, a2, ...])"""
    return sum(c * (x**i) for i, c in enumerate(poly))

def poly_integral(poly):
    """다항식 poly의 부정적분 (상수항 0) 반환: [a0, a1, ...] -> [0, a0, a1/2, ...]"""
    return [0] + [poly[i] / (i+1) for i in range(len(poly))]

def poly_add(poly1, poly2):
    """두 다항식의 덧셈: poly1 + poly2"""
    n = max(len(poly1), len(poly2))
    res = [0]*n
    for i in range(n):
        a = poly1[i] if i < len(poly1) else 0
        b = poly2[i] if i < len(poly2) else 0
        res[i] = a + b
    return res

def poly_scale(poly, scalar):
    """다항식 poly의 모든 계수를 scalar배 함"""
    return [scalar * c for c in poly]

def poly_subtract(poly1, poly2):
    """두 다항식의 뺄셈: poly1 - poly2"""
    return poly_add(poly1, poly_scale(poly2, -1))

def poly_mul(poly1, poly2):
    """두 다항식의 곱 (합성곱)"""
    res = [0]*(len(poly1) + len(poly2) - 1)
    for i, a in enumerate(poly1):
        for j, b in enumerate(poly2):
            res[i+j] += a * b
    return res

def poly_compose(poly, shift):
    """
    다항식 poly(x)를 x -> (z - shift)로 치환하여 Q(z)=poly(z-shift)를 구함.
    poly: 계수 리스트 [a0, a1, ...]
    반환: Q(z)를 나타내는 계수 리스트.
    """
    n = len(poly)
    result = [0]*n
    for i, a in enumerate(poly):
        for j in range(i+1):
            result[j] += a * comb(i, j) * ((-shift)**(i-j))
    return result

# 간단한 테스트
print("Cell 1 테스트:")
poly_example = [2, 3]  # 2 + 3x
print("poly_example =", poly_example)
print("eval_poly(poly_example, 5) =", eval_poly(poly_example, 5))
print("poly_integral(poly_example) =", poly_integral(poly_example))
print("poly_compose(poly_example, 2) =", poly_compose(poly_example, 2))
```

### 4. piecewise 함수와 uniform 구간의 합성곱

```python
def convolve_piecewise_with_uniform(piecewise_segments, uniform_seg):
    """
    piecewise_segments: 리스트, 각 원소는 Segment (정의역: [a1, b1], 다항식 poly)
    uniform_seg: 단일 uniform Segment (정의역: [a2, b2], poly = [c2])
    반환: 두 함수의 합성곱 결과를 나타내는 새로운 Segment 리스트.
    
    여기서는 seg.poly가 상수 혹은 1차 함수인 경우로 한정.
    """
    new_segments = []
    c2 = uniform_seg.poly[0]
    a2, b2 = uniform_seg.start, uniform_seg.end
    for seg in piecewise_segments:
        a1, b1, poly1 = seg.start, seg.end, seg.poly
        # F(x): antiderivative of poly1
        F = poly_integral(poly1)
        
        # breakpoints:
        z_low = a1 + a2
        z_mid1 = b1 + a2  # when upper limit changes from z - a2 to b1
        z_mid2 = a1 + b2  # when lower limit changes from a1 to z - b2
        z_high = b1 + b2
        
        # 가정: z_low < z_mid1 < z_mid2 < z_high (주로 b1-a1 < b2-a2 인 경우)
        # Region 1: z in [z_low, z_mid1]: 
        #   L(z)=a1 (상수), U(z)=z - a2.
        #   f(z) = c2 * (F(z - a2) - F(a1))
        #   다항식: poly_region1 = c2 * ( poly_compose(F, a2) - F(a1) )
        F_comp1 = poly_compose(F, a2)  # F(z - a2) as polynomial in z
        const_Fa1 = F[0] + 0  # F(a1) evaluated: since F는 다항식, 직접 평가:
        F_a1 = eval_poly(F, a1)
        poly_region1 = poly_subtract(F_comp1, [F_a1])
        poly_region1 = poly_scale(poly_region1, c2)
        new_segments.append(Segment(z_low, z_mid1, poly_region1))
        
        # Region 2: z in [z_mid1, z_mid2]:
        #   L(z)=a1, U(z)=b1, 상수 값: f(z) = c2*(F(b1)-F(a1))
        const_val = c2 * (eval_poly(F, b1) - eval_poly(F, a1))
        new_segments.append(Segment(z_mid1, z_mid2, [const_val]))
        
        # Region 3: z in [z_mid2, z_high]:
        #   L(z)=z - b2, U(z)=b1, f(z) = c2*(F(b1)-F(z - b2))
        F_comp2 = poly_compose(F, b2)  # F(z - b2)
        poly_region3 = poly_subtract([eval_poly(F, b1)], F_comp2)
        poly_region3 = poly_scale(poly_region3, c2)
        new_segments.append(Segment(z_mid2, z_high, poly_region3))
    return new_segments

# 테스트: convolve_piecewise_with_uniform
print("4 테스트:")
# 테스트로, convolve_two_uniform으로 만든 piecewise function을 uniform 구간과 합성곱
pw_test = test_conv  # from Cell 3
# 예를 들어, uniform_seg로 left_intervals[0] 사용 (비록 이미 uniform이지만)
conv_pw = convolve_piecewise_with_uniform(pw_test, left_intervals[0])
for seg in conv_pw:
    print(seg)
print("정말 감사합니다.\n")
```

### 5. piecewise PDF 병합 및 평행이동 함수

```python
def merge_piecewise_pdf(pw1, pw2):
    """
    두 piecewise PDF (Segment 리스트)를 병합.
    각 Segment의 구간이 다를 수 있으므로, 모든 경계점을 모아 작은 구간으로 나누고,
    해당 구간에서 두 함수의 값을 (다항식) 합산합니다.
    """
    # 모든 경계점 수집
    boundaries = set()
    for seg in pw1 + pw2:
        boundaries.add(seg.start)
        boundaries.add(seg.end)
    boundaries = sorted(boundaries)
    
    merged = []
    # 각 작은 구간에 대해
    for i in range(len(boundaries)-1):
        a = boundaries[i]
        b = boundaries[i+1]
        # pw1의 기여
        poly_sum = [0]
        for seg in pw1:
            if seg.start <= a and seg.end >= b:
                poly_sum = poly_add(poly_sum, seg.poly)
        # pw2의 기여
        for seg in pw2:
            if seg.start <= a and seg.end >= b:
                poly_sum = poly_add(poly_sum, seg.poly)
        merged.append(Segment(a, b, poly_sum))
    return merged

def shift_piecewise_pdf(pw, shift):
    """
    piecewise PDF pw를 x -> x - shift로 변환.
    즉, f_new(x) = f_old(x - shift).
    각 Segment의 구간은 [start+shift, end+shift]가 되고,
    다항식은 poly(x - shift) = poly_compose(poly, shift)로 변환.
    """
    shifted = []
    for seg in pw:
        new_start = seg.start + shift
        new_end = seg.end + shift
        new_poly = poly_compose(seg.poly, shift)
        shifted.append(Segment(new_start, new_end, new_poly))
    return shifted

# 간단한 테스트:
print("Cell 5 테스트: merge 및 shift")
pw1 = [Segment(0, 1, [1]), Segment(1, 2, [2])]
pw2 = [Segment(0.5, 1.5, [3])]
merged = merge_piecewise_pdf(pw1, pw2)
print("Merged:")
for seg in merged:
    print(seg)
shifted = shift_piecewise_pdf(pw1, 2)
print("Shifted (by 2):")
for seg in shifted:
    print(seg)
print("정말 감사합니다.\n")
```

### 6. 4개 구간에 대해 반복 합성곱 수행 함수

```python
def convolve_segments_list(segments_list):
    """
    segments_list: 리스트, 각 원소가 Segment (PDF)
    4개 구간에 대해 for문으로 합성곱 수행.
    초기 piecewise PDF는 첫 번째 Segment로 시작하여,
    이후 나머지 구간에 대해 convolve_piecewise_with_uniform를 적용.
    """
    piecewise = [segments_list[0]]
    for seg in segments_list[1:]:
        piecewise = convolve_piecewise_with_uniform(piecewise, seg)
        # 필요 시, 병합:
        piecewise = merge_piecewise_pdf(piecewise, [])
    return piecewise

# 테스트: 2개 구간의 합성곱 (예: left_intervals[0]와 left_intervals[1])
print("Cell 6 테스트:")
test_segments = [left_intervals[0], left_intervals[1]]
conv_result = convolve_segments_list(test_segments)
for seg in conv_result:
    print(seg)
print("감사합니다.\n")
```

### 7. 케이스별 조합(pdf 계산) 함수 구현

```python
import itertools

def compute_case_pdf(left_count, right_count):
    """
    left_count: 좌측 구간에서 선택할 개수
    right_count: 우측 구간에서 선택할 개수 (총합 4)
    반환: 해당 케이스의 누적 piecewise PDF와 조합 수.
    """
    case_pdf = None  # 누적 piecewise PDF
    comb_count = 0
    for left_choice in itertools.combinations(left_intervals, left_count):
        for right_choice in itertools.combinations(right_intervals, right_count):
            comb_count += 1
            selected = list(left_choice) + list(right_choice)
            pdf_piece = convolve_segments_list(selected)
            if case_pdf is None:
                case_pdf = pdf_piece
            else:
                case_pdf = merge_piecewise_pdf(case_pdf, pdf_piece)
    return case_pdf, comb_count

# 간단한 테스트: 케이스 (3,1)
print("Cell 7 테스트: 케이스 (3,1)")
pdf_case, count_case = compute_case_pdf(3, 1)
print("조합 수:", count_case)
for seg in pdf_case:
    print(seg)
print("정말 감사합니다.\n")
```

### 8. 전체 PDF 및 CDF, 구간 확률 함수 구현

```python
def eval_integral_poly(poly, x):
    """다항식 poly의 부정적분을 x에서 평가 (적분 상수는 0로 가정)"""
    integ_poly = poly_integral(poly)
    return eval_poly(integ_poly, x)

def compute_cdf(x, piecewise_pdf):
    """
    piecewise_pdf: Segment 리스트
    x: 평가 위치
    반환: CDF(x)
    """
    total = 0
    for seg in piecewise_pdf:
        if x >= seg.end:
            total += eval_integral_poly(seg.poly, seg.end) - eval_integral_poly(seg.poly, seg.start)
        elif x > seg.start:
            total += eval_integral_poly(seg.poly, x) - eval_integral_poly(seg.poly, seg.start)
    return total

def probability_between(a, b, piecewise_pdf):
    """구간 [a, b] 사이의 확률"""
    return compute_cdf(b, piecewise_pdf) - compute_cdf(a, piecewise_pdf)

# 간단한 테스트:
print("Cell 8 테스트: CDF 및 구간 확률")
# 임의의 piecewise_pdf: 단일 구간 [0,1]에서 상수 2
test_pdf = [Segment(0, 1, [2])]
print("CDF(0.5) =", compute_cdf(0.5, test_pdf))
print("P(0.2,0.8) =", probability_between(0.2, 0.8, test_pdf))
print("지금까지 pdf -> cdf까지. Okay..\n")
```

```python
# 케이스: (좌측/우측 선택 개수)
cases = [(4,0), (3,1), (2,2), (1,3), (0,4)]
case_results = []
total_comb = 0

print("Cell 9: 각 케이스별 PDF 계산 진행 중...")
for lc, rc in cases:
    pdf_case, count_case = compute_case_pdf(lc, rc)
    print(f"케이스 ({lc}/{rc}) 조합 수:", count_case)
    case_results.append((pdf_case, count_case))
    total_comb += count_case
print("전체 조합 수:", total_comb)

# 전체 PDF = 각 케이스 PDF에 조합 수 가중치를 곱해 합산한 후 전체 조합 수로 나눔.
# 각 케이스에 대해 평행이동 및 스케일링(평균: 합/4, 즉 f_Y(y)=4 f_X(4y)) 필요.
# 여기서는 간단히, 케이스 PDF를 가중합하는 예시를 보입니다.
def scale_pdf(piecewise, scale):
    """X -> X/scale 변환: f_Y(y) = scale * f_X(scale*y)"""
    scaled = []
    for seg in piecewise:
        new_start = seg.start / scale
        new_end = seg.end / scale
        # 새로운 다항식: scale * poly evaluated at scale*y = poly_compose(poly, 0)? 
        # (여기서는 단순히 계수를 scale배 함)
        new_poly = poly_scale(seg.poly, scale)
        scaled.append(Segment(new_start, new_end, new_poly))
    return scaled

def add_case_pdfs(case_results, total_comb):
    total_pdf = None
    for pdf_case, count_case in case_results:
        # 예: 평행이동 및 스케일링 처리는 필요시 shift_piecewise_pdf 등을 적용
        # 여기서는 그대로 사용
        if total_pdf is None:
            total_pdf = poly_scale_pdf(pdf_case, count_case)
        else:
            total_pdf = merge_piecewise_pdf(total_pdf, poly_scale_pdf(pdf_case, count_case))
    # 전체 PDF에 대해 조합 수로 나눔
    # 즉, 각 다항식의 계수를 (1/total_comb) 배
    return poly_scale_pdf(total_pdf, 1/total_comb)

def poly_scale_pdf(pw, scalar):
    """각 Segment의 다항식에 scalar 배 적용"""
    new_pw = []
    for seg in pw:
        new_pw.append(Segment(seg.start, seg.end, poly_scale(seg.poly, scalar)))
    return new_pw

total_pdf = add_case_pdfs(case_results, total_comb)

# 최종 PDF를 평균으로 스케일링: Y = X/4, f_Y(y) = 4 * f_X(4y)
final_pdf = scale_pdf(total_pdf, 4)

print("최종 PDF (piecewise) 출력:")
for seg in final_pdf:
    print(seg)

# 최종 CDF 및 특정 구간 확률 계산 예시
a_val, b_val = -1, 1
prob_interval = probability_between(a_val, b_val, final_pdf)
print(f"구간 [{a_val}, {b_val}]의 확률:", prob_interval)
```
